```python
# import konlpy and other libraries to analyse Korean text data
from konlpy.tag import Kkma
from konlpy.tag import Okt

# create an instance of Kkma and Okt
kkma = Kkma()
okt = Okt()
```

```python
# import text data from ../raw-data/ directory
os.getcwd()
data_dir = os.getcwd() + '/drive/MyDrive/aiffel/dlthon/raw-data'
print(data_dir)
```

```python
train = pd.read_csv(data_dir + '/train.csv')
train
```

```python
# 개행 문자 대체하기
train['conversation'] = train['conversation'].apply(lambda x: x.replace('\n', ' '))
train.head()
```

```python
# 형태소 칼럼 만들기
train['morphs'] = train['conversation'].apply(lambda x: kkma.morphs(x))
train.head(10)
```

```python
# 명사 칼럼 만들기
train['nouns'] = train['conversation'].apply(lambda x: kkma.nouns(x))
train.head(10)
```

```python
# 데이터에서 괄호와 따옴표와 스페이스를 제거한다.
train['morphs'] = train['morphs'].apply(lambda x: str(x).replace('[', "").replace(']', "").replace("'", "").replace(" ", ""))
train['nouns'] = train['nouns'].apply(lambda x: str(x).replace('[', "").replace(']', "").replace("'", "").replace(" ", ""))
```
