## 전체 형태소 내용 분석

## 클래스별 단어와 형태소를 분석해보자.
각 클래스에서 빈출되는 단어를 기반으로 일반 데이터를 만들면 모델의 과적합을 줄일 수 있다고 생각한다.

```python
categories = {
    "협박 대화": 0, # threats
    "갈취 대화": 1, # extortion
    "직장 내 괴롭힘 대화": 2, # harassment at works
    "기타 괴롭힘 대화": 3, # other harassment
    "일반 대화": 4 # normal conversations
}
```

```python
threat_dataset = train[train['class'] == '협박 대화']
extortion_dataset = train[train['class'] == '갈취 대화']
harassment_at_works_dataset = train[train['class'] == '직장 내 괴롭힘 대화']
other_harassment_dataset = train[train['class'] == '기타 괴롭힘 대화']
```

```python
# 클래스별로
def get_morph_tokens(dataset):
    # 각 레코드별 토큰
    morphs = []
    for sentence in dataset['morphs']:
        morphs.append(sentence.split(','))

    # 모든 레코드의 토큰을 하나의 리스트에 저장
    morph_tokens = []
    for sentence in morphs:
        for word in sentence:
            morph_tokens.append(word)
    return morph_tokens

def get_noun_tokens(dataset):

    # 각 레코드별 토큰
    nouns = []
    for sentence in dataset['nouns']:
        nouns.append(sentence.split(','))

    # 모든 레코드의 토큰을 하나의 리스트에 저장
    noun_tokens = []
    for sentence in nouns:
        for word in sentence:
            noun_tokens.append(word)
    return noun_tokens
```

```python
threat_tokens = get_noun_tokens(threat_dataset)
extortion_tokens = get_noun_tokens(extortion_dataset)
harassment_at_works_tokens = get_noun_tokens(harassment_at_works_dataset)
other_harassment_tokens = get_noun_tokens(other_harassment_dataset)
```

```python
threat_counters = Counter(threat_tokens)
extortion_counters = Counter(extortion_tokens)
harassment_at_works_counters = Counter(harassment_at_works_tokens)
other_harassment_counters = Counter(other_harassment_tokens)
harassment_at_works_counters.most_common(500)
```
